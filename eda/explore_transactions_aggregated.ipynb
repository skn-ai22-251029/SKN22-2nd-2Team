```python
import matplotlib.pyplot as plt
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
```

```python
tx_agg = pd.read_parquet("../data/processed/transactions_aggregated.parquet")
```

```python
tx_agg.head(30)
```

```python
tx_agg.describe()
```

# STEP 1. 상태 기반 피처 검증 (가장 중요)

## 1-1. days_since_last_payment 분포
- 음수 / 이상치 존재 여부
- churn 직전 상태로서 합리적인 분포인지

```python
plt.figure(figsize=(6,4))
plt.hist(tx_agg["days_since_last_payment"], bins=50)
plt.title("days_since_last_payment distribution")
plt.xlabel("days")
plt.ylabel("count")
plt.show()
```

## 1-2. days_since_last_cancel 결측 구조
- 결측 = 취소 이력 없음 가정이 성립하는지

```python
missing_ratio = tx_agg["days_since_last_cancel"].isna().mean()
missing_ratio
```

## 1-3. is_auto_renew_last 값 검증
- binary 정상 여부
- 이상한 코드값 존재 여부

```python
tx_agg["is_auto_renew_last"].value_counts().plot(kind="bar")
plt.title("is_auto_renew_last value counts")
plt.show()
```

## 1-4. last_plan_days 분포
- 플랜 기간이 정상적으로 집계됐는지

```python
tx_agg["last_plan_days"].value_counts().sort_index().plot(kind="bar")
plt.title("last_plan_days distribution")
plt.show()
```

# STEP 2. 누적 히스토리 피처 검증

## 2-1. total_payment_count 분포
- 0값 존재 여부 (집계 오류 가능성)
- long-tail 구조 확인

```python
plt.figure(figsize=(6,4))
plt.hist(tx_agg["total_payment_count"])
plt.title("total_payment_count distribution")
plt.show()
```

## 2-2. total_amount_paid 분포 (로그 스케일)
- 과도한 outlier 확인
- 금액 누적이 합리적인지

```python
plt.figure(figsize=(6,4))
plt.hist(tx_agg["total_amount_paid"], bins=50)
plt.yscale("log")
plt.title("total_amount_paid (log scale)")
plt.show()
```

## 2-3. avg_amount_per_payment sanity
- total / count 계산 일관성 확인

```python
plt.figure(figsize=(6,4))
plt.hist(tx_agg["avg_amount_per_payment"], bins=50)
plt.title("avg_amount_per_payment distribution")
plt.show()
```

## 2-4. unique_plan_count 분포
- 비정상적으로 많은 플랜 수 확인
- cap 필요 여부 판단

```python
tx_agg["unique_plan_count"].value_counts().sort_index().plot(kind="bar")
plt.title("unique_plan_count distribution")
plt.show()
```

## 2-5. subscription_months_est 분포
- 0 또는 음수 존재 여부
- 과도한 값 확인

```python
plt.figure(figsize=(6,4))
plt.hist(tx_agg["subscription_months_est"], bins=50)
plt.title("subscription_months_est distribution")
plt.show()
```

# STEP 3. Recency 피처 검증 (형식적 검증)

## 3-1. payment_count_last_30d
- 대부분 0인지 확인 (의도된 sparsity)

```python
tx_agg["payment_count_last_30d"].value_counts().sort_index().plot(kind="bar")
plt.title("payment_count_last_30d")
plt.show()
```

## 3-2. w30 vs w90 정보 증가 확인
- w90이 실제로 추가 정보 제공하는지
- 전부 y=x면 w90 의미 없음

```python
plt.figure(figsize=(5,5))
plt.scatter(
    tx_agg["payment_count_last_30d"],
    tx_agg["payment_count_last_90d"],
    s=5
)
plt.xlabel("payment_count_last_30d")
plt.ylabel("payment_count_last_90d")
plt.title("w30 vs w90 payment count")
plt.show()
```
